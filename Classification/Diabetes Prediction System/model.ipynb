"Body Mass Index" (BMI): Which is a measure calculated using a person's weight and height to determine if they are underweight, normal weight, overweight, or obese.

"Systolic Blood Pressure" (SBP): which measures the pressure in your arteries when your heart beats. High blood pressure can be linked to diabetes.

"Diastolic Blood Pressure"(DBP): which is the pressure in your arteries when your heart rests between beats.

"Fasting Plasma Glucose"(FPG) : which measures blood sugar levels after not eating for at least 8 hours and is used to check for diabetes. 

"Cholesterol" (Chol) : It indicates the amount of fat in the blood, which helps doctors understand heart health.

"Tri-ponderal Mass Index"(Tri) : a measure of body fat calculated by dividing weight (in kilograms) by height (in meters) cubed (kg/mÂ³). In diabetes studies, TMI is used to assess body fat more accurately than BMI, especially in children and adolescents.

"High-Density Lipoprotein" (HDL) :HDL often called "good" cholesterol. In a diabetes dataset, HDL levels help assess heart health, as higher HDL can lower heart disease risk. 

"Low-Density Lipoprotein"(LDL) : LDL often called "bad cholesterol" because high levels can lead to heart disease.  In diabetes studies, LDL levels are tracked to assess heart health risks.

"Alanine Aminotransferase" (ALT) : an enzyme mainly found in the liver. In a diabetes dataset, ALT levels can indicate liver health, as elevated levels may suggest liver damage or disease. 

"Blood Urea Nitrogen" (BUN):BUN levels help assess kidney function, which is important because diabetes can affect how well kidneys work. 

"Creatinine Clearance Rate" (CCR):measures how well the kidneys filter waste from the blood.CCR helps assess kidney function.

"First Fasting Plasma Glucose" (FFPG): Which is the blood sugar level measured after fasting, typically during the first trimester of pregnancy, to assess the risk of developing gestational diabetes mellitus (GDM).

```python
def convert_smoking(data):
    if(data==1.0):
        return "Never Smoked"
    elif(data==2.0):
        return "Once up on a time"
    elif(data==3.0):
        return "Occasional Smoker"
    else:
        return "Daily Smoker"
```

```python
dataset['smoking']=dataset['smoking'].apply(convert_smoking)
```

```python
np.unique(dataset['smoking'])
```

```python
np.unique(dataset['drinking'])
```

```python
def convert_drinking(data):
    if data<=1.0 and data>=0.0:
        return "Never Drank"
    elif data<=2.0 and data>1.0:
        return "Once up on a time"
    elif data<=3.0 and data>2.0:
        return "Occasional Drinker"
    else:
        return "Daily Drinker"
```

```python
dataset['drinking']=dataset['drinking'].apply(convert_drinking)
```

```python
X.drop(columns=['Gender'],inplace=True)
```

```python
X.columns
```

```python
X=X.drop(columns=['ALT','CCR'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
column=['BMI','SBP','DBP','FPG']
ss=StandardScaler()
x_train[column]=ss.fit_transform(x_train[column])
x_test[column]=ss.transform(x_test[column])
```

```python
x_train.columns
```

```python
x_train.head(2)
```

# Model Seclection

```python
models={}
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
Lr=LogisticRegression()
Lr.fit(x_train,y_train)
y_pred_Lr=Lr.predict(x_test)
```

```python
models[type(Lr).__name__]=f"{accuracy_score(y_test,y_pred_Lr):.3f}"
```

```python
from sklearn.tree import DecisionTreeClassifier
Dt=DecisionTreeClassifier()
Dt.fit(x_train,y_train)
y_pred_Dt=Dt.predict(x_test)
```

```python
models[type(Dt).__name__]=f"{accuracy_score(y_test,y_pred_Dt):.3f}"
```

```python
from sklearn.ensemble import RandomForestClassifier
Rf=RandomForestClassifier(n_estimators=100)
Rf.fit(x_train,y_train)
y_pred_Rf=Rf.predict(x_test)
```

```python
models[type(Rf).__name__]=f"{accuracy_score(y_test,y_pred_Rf):.3f}"
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='rbf')
svc.fit(x_train,y_train)
y_pred_svc=svc.predict(x_test)
```

```python
models[type(svc).__name__]=f"{accuracy_score(y_test,y_pred_svc):.3f}"
```

```python
from sklearn.svm import SVC
svcl=SVC(kernel='linear')
svcl.fit(x_train,y_train)
y_pred_svcl=svc.predict(x_test)
```

```python
models['scv_linear']=f"{accuracy_score(y_test,y_pred_svcl):.3f}"
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred_nb=nb.predict(x_test)
```

```python
models[type(nb).__name__]=f"{accuracy_score(y_test,y_pred_nb):.3f}"
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
```

```python
models[type(knn).__name__]=f"{accuracy_score(y_test,y_pred_knn):.3f}"
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
y_pred_xgb=xgb.predict(x_test)
```

```python
models['XGBoost']=f"{accuracy_score(y_test,y_pred_xgb):.3f}"
```
