```python
def null_values(dataset):
    n=[i for i in dataset.describe().columns]
    c=[i for i in dataset.columns if i not in dataset.describe().columns]
    for i in n:
        dataset[i].fillna(dataset[i].median(),inplace=True)
    for i in c:
        dataset[i].fillna(dataset[i].value_counts().index[0],inplace=True)
    return dataset,n,c
```

```python
data,numerical_data,categorical_data=null_values(data)
print("Numerical Columns : ",numerical_data)
print("-----------------------------------------------------------------------")
print("Categorical columns : ",categorical_data)
print("-----------------------------------------------------------------------")
data.isna().sum()
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=10,random_state=0)
kmeans.fit(data)
```

```python
# Finding the WCSS score
WCSS=[]
for i in range(2,15):
    wcss=KMeans(n_clusters=i,init="k-means++",random_state=0)
    wcss.fit(wcss1)
    WCSS.append(wcss.inertia_)
```

```python
plt.plot(range(2,15),WCSS,color="green")
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```
